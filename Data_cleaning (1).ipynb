```python
df.price = df.price.str.strip().str.replace(',','').str.replace('₹','').astype('float64')
```

```python
df.head()
```

```python
df['duration'] = df.tax_charges.apply(lambda x: 3 if re.findall('3 nights', x) else np.nan)
```

```python
df['persons'] = df.tax_charges.apply(lambda x: 2 if re.findall('2 adults', x) else np.nan)
```

```python
df.features.sample(500)
```

```python
df
```

```python
df['cancellation_policy'] = df['features'].apply(lambda x: 'Yes' if re.findall(r'(?:Free cancellation)+', x) else 'No')
```

```python
df.cancellation_policy.value_counts()
```

```python
df['pre_payment'] = df['features'].apply(lambda x: 'No' if re.findall(r'(?:No prepayment needed)+', x) else 'Yes')
```

```python
df.pre_payment.value_counts()
```

```python
df['breakfast_included'] = df['features'].apply(lambda x: 'Yes' if re.findall(r'(?:Breakfast included)', x) else 'No')
```

```python
df.breakfast_included.value_counts()
```

```python
df.reset_index(inplace=True)
```

```python
df.reviews = df.reviews.apply(lambda x: str(x).replace('external ', '').strip() if 'external' in str(x) else str(x).strip())
```

```python
df.reviews = df.reviews.apply(lambda x: str(x).replace(',', '') if ',' in str(x) else x).astype('float64')
```
